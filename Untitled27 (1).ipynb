```python
import pandas as pd

df = pd.read_excel(
    "assignment/Association Rules/Online retail.xlsx",
    header=None
)

df.head()
```

```python
transactions = []

for i in range(len(df)):
    transactions.append(df.iloc[i,0].split(','))
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_items = apriori(
    basket,
    min_support=0.01,
    use_colnames=True
)

print(frequent_items.head())
```

```python
rules = association_rules(
    frequent_items,
    metric="confidence",
    min_threshold=0.3
)

print(rules[['antecedents',
             'consequents',
             'support',
             'confidence',
             'lift']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x="support",
    y="confidence",
    size="lift",
    data=rules
)

plt.title("Association Rules")
plt.show()
```

Association Rule Mining was performed using the Apriori algorithm.
Frequent itemsets were generated based on minimum support.
Association rules were extracted using confidence metric.
The generated rules help identify relationships between products
and can be used for market basket analysis.
